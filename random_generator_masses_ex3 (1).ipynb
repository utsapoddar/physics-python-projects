This code generates the mass of the rocket (without fuel) and the initial total mass of the rocket plus fuel, i.e. when the fuel tank of the rocket is 100% filled prior to launching.
```python
import random

# seed input
seed = int(input('Enter ANY integer number from your head: '))
print('You entered as seed for the random number generator:', seed)

# seed to initialize random number generator
random.seed(seed)

# the mass of the rocket without fuel will be between 4000 Kg and 4500 Kg
Mr_min = 4000
Mr_max = 4500

mass_rocket = random.randint(Mr_min, Mr_max)
print('The mass of the rocket (without fuel) is:', mass_rocket, 'Kg')

# the mass of fuel that the tank can carry will be between 8800 Kg and 9000 Kg
Mf_min = 8800
Mf_max = 9000

mass_fuel = random.randint(Mf_min, Mf_max)
print('The mass of fuel in the tank is:', mass_fuel, 'Kg')

# total mass of the rocket with tank full (in Kg)
total_mass = mass_rocket + mass_fuel
print('The total mass of the rocket with tank full is:', total_mass, 'Kg')
```
