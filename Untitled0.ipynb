```
import matplotlib.pyplot as plt
import math
import time

#setting parameters

#gravitational accelaration 
g=9.81
#diameter of the object
D=10**-4
#volume of the object
vol=(4/3)*math.pi* (D/2)**3
#density of the object
rho=2000
# mass of the object
m=rho*vol

#features of the drag constant
B=1.6*10**-4
b=B*D
#time spacing
dt=0.001

#angle of throw (in degree)
theta=10
#degree to radians
theta_rads=math.radians(theta)

#initial conditions (with air resistance)
t=0
v0=2.3
vy=v0*math.sin(theta_rads)
vx=v0*math.cos(theta_rads)
y=0
x=0

#initial conditions (in vacuum)
vy_vac=vy
vx_vac=vx
y_vac=y
x_vac=x

#lists for the motion in air resistence
timev=[t]
velocityy=[vy]
velocityx=[vx]
yy=[y]
xx=[x]

#lists for the motion in vacuum
velocityy_vac=[vy]
velocityx_vac=[vx]
yy_vac=[y]
xx_vac=[x]

#loop for the Euler integration
while(y_vac>=0):
  #model with air resistence
  dvy=-g*dt-(b/m)*vy*dt
  vy=vy+dvy

  
  dy=dt*vy
  y=y+dy

  dvx=-(b/m)*vx*dt
  vx=dvx+vx

  dx=dt*vx
  x=x+dx

  t=t+dt

  if y<0:
    x=x-dx
    y=0.0
    vy=0.0
    vx=0.0

  timev.append(t)
  velocityy.append(vy)
  velocityx.append(vx)
  yy.append(y)
  xx.append(x)

  dvy_vac=-g*dt
  vy_vac=vy_vac+dvy_vac
  dy_vac=dt*vy_vac
  y_vac=y_vac+dy_vac

  dvx_vac=0
  vx_vac=vx_vac+dvx_vac
  dx_vac=dt*vx_vac
  x_vac=x_vac+dx_vac

  velocityy_vac.append(vy_vac)
  velocityx_vac.append(vx_vac)
  yy_vac.append(y_vac)
  xx_vac.append(x_vac)

plt.plot(xx,yy,'b',label="In air")
plt.plot(xx_vac,yy_vac,'r',label="In vacuum")
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()
plt.show()
```
