```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# defining the initial system (aka [x, y, z] positions in space)
initial_state=[0.1,0,0]

#defining the system parameters sigma, rho and beta
sigma=10
rho=28
beta=8/3

#the range of values of t for which the graphs are to be plotted
t=np.linspace(0,100,10000)

#provided set of non-linear system of equations
def deriv(param,t,sigma,rho,beta):
  #parameter defining the values of x, y and z
  x,y,z=param
  #formulae for the system of equations
  dxdt=sigma*(y-x)
  dydt=x*(rho-z)-y
  dzdt=x*y-(beta*z)

  return dxdt,dydt,dzdt
#parameter that stores the initial value of x, y and z
param0=initial_state[0],initial_state[1],initial_state[2]

#using odeint to determine the solution
result=odeint(deriv,param0,t,args=(sigma,rho,beta))

#seperating the solution for x, y and z
x,y,z=result.T

#ploting the figures and setting the size to miminc the graphs in the final
fig,axs=plt.subplots(1,3,figsize=(18,6))

#ploting the x-y phase plane, and setting the linewidth to mimic given graphs
axs[0].plot(x,y,'r',linewidth=0.2)
#providing an appropriate title
axs[0].set_title("x-y phase plane")
#ploting the x-z phase plane, and setting the linewidth to mimic given graphs
axs[1].plot(x,z,'purple',linewidth=0.2)
#providing an appropriate title
axs[1].set_title("x-z phase plane")
#ploting the y-z phase plane, and setting the linewidth to mimic given graphs
axs[2].plot(y,z,'b',linewidth=0.2)
#providing an appropriate title
axs[2].set_title("y-z phase plane")

plt.show()

```

Here the code started off by importing the required libraries and adding the lines for initial state and defining the parameters of the system. The function for the derivatives was defined using the derivatives provided in the final assignment. Then a parameter was introduced which stored all the values of the initial state, and using odeint from scipy.integrate the solution for the derivates was determined in a variable called result. However, result contained the solution for x, y and z, hence needed to be separated. Lastly the x-y, x-z and y-z plots are plotted to mimic the pictures provided in the final assignment.
