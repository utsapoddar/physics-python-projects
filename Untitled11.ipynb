```python
import matplotlib.pylab as plt
import numpy as np
import sys
sys.setrecursionlimit(1500)
itterations = 1000 # Number of itterations to compute
itter_range = range(itterations)
print(range(1,len(itter_range)))

N = 100 # Number of divisions in x
T_r=109.2
T_l=16.9
L=8
x = np.linspace(0, L, N)
T_list = np.zeros(N, dtype=np.double)
T_list[-1] = T_r
T_list[0]=T_l
T_select = [1,50, 70, 100,500, itterations - 1]
select = 0

# define a function to calculate the interior difference and replace the interior value
def interior_difference(T_list):
  T_list[1:-1] = (T_list[0:-2] + T_list[2:])/2

#def recursive(itter_range,T_list,T_select,select):
  #for i in itter_range:
    #interior_difference(T_list)
    #if i == T_select[select]:
      #plt.plot(x, T_list, label="Itteration: %d" % (i))
      #select += 1
def r2(itterations,select,T_select,T_list,i=0):
  if itterations==i:
    return ("done")
  if itterations==T_select[select]:
    plt.plot(x, T_list, label="Itteration: %d" % (i))
    select += 1
    interior_difference(T_list)
    return r2(itterations-1,select,T_select,T_list,i=0)
  else:
    interior_difference(T_list)
    return r2(itterations-1,select,T_select,T_list,i=0)

# apply interior difference function 10000 times and plot the current curve at select values of the itteration
#for i in itter_range:
  #interior_difference(T_list)
  #if i == T_select[select]:
    #plt.plot(x, T_list, label="Itteration: %d" % (i))
    #select += 1
r2(itterations,select,T_select,T_list,i=0)
#plt.plot(x, ((T_r-T_l)/L)*x+T_l, label="Analytic Solution")
plt.title("1D Numeric Solution To The Laplace Equation")
plt.xlabel("Distance (m)")
plt.ylabel("Temperature (C)")
plt.legend()
plt.show()

epsilon_list=[]
for i in range(len(x)):
  epsilon=T_list[i]-((T_r-T_l)/L)*(float(x[i]))+T_l
  epsilon_list.append(epsilon)

plt.plot(x, epsilon_list, label="error")
plt.title("error vs x")
plt.xlabel("x (m)")
plt.ylabel("error (C)")
plt.legend()


# The numeric solution is remarkably close to the analytic solution and would likely continue to approach the analytic solution with more itterations.
# Visually, the numeric solution could not be distinguished from the analytic solution without contrasting them.

plt.show()
```
