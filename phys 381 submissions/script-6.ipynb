````python
import matplotlib.pylab as plt
import numpy as np
from sympy import *
#creating a symbol x
x_1=Symbol('x_1')
f=a*x_1*x_1+b*x_1+c
f_prime=f.diff(x_1)
f=lambdify(x_1,f)
f_prime=lambdify(x_1,f_prime)
#taking input from the user for a single guess value 
x_1=float(input("Enter your guess value: "))
#making arrays for the data points to be stored into
x_values=[]
f_values=[]
#ploting the initial point 
plt.plot(x_1,f(x_1),'ro')
#initializing the counter for the number of tests
nsteps=0
print("The values of x and f(x) BEFORE the absolute value of f(x) becomes less than tolerance are %.2f and %.2f" %(x_1,f(x_1)))
#loop for determining the root using the NR method
while abs(a*x_1*x_1+b*x_1+c)>tol:
  #the formula for NR method
  x_1=x_1-(f(x_1)/f_prime(x_1))
  #determining the f(x) value for each iteration of the point 
  f_value=a*x_1*x_1+b*x_1+c
  #adding the values to the appropriate arrays 
  x_values.append(x_1)
  f_values.append(f_value)
  #counting the number steps using the loop 
  nsteps=nsteps+1
#printing the values as instructed
print("The values of x and f(x) AFTER the absolute value of f(x) becomes less than tolerance are %.2f and %.2f" %(x_1,f(x_1)))
print("The number of steps taken to determine the root is ",nsteps)
#ploting the converged point on the parabola
plt.plot(x_1,f_value,'ro')
#plotting the graph of f(x) vs x 
x=np.arange(-4.0,6.0,0.2)
plt.plot(x,a*x*x+b*x+c)
plt.plot(x,0.0*x)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()


````
